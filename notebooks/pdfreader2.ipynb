## pdf reader with pymupdf4llm

```python
import pymupdf4llm

reader = pymupdf4llm.to_markdown("Annual-Report202425.pdf",ignore_images= True,page_chunks=True)
```

```python
print(reader)
# now work with the markdown text, e.g. store as a UTF8-encoded file
import pathlib
pathlib.Path("output.md").write_bytes(reader.encode())
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(chunk_size=1000,chunk_overlap=200)

all_splits = text_splitter.create_documents([reader])
```

```python
all_splits
```

```python
import json 
with open("C:\\krishisahayi\\lang-chain-bot\\data\\scraped_chunks.json", "a", encoding="utf-8") as f:
    json.dump(
        [{"page_content": doc.page_content, "metadata": doc.metadata} for doc in all_splits],
        f,
        ensure_ascii=False,
        indent=2
    )
```
